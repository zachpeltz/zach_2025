### What This Program Does:
- This program demonstrates a functionality where a user inputs a list of numbers, then the program calculates the factorial of each number, and then visualizes the results using a bar chart.

```python
import math
import matplotlib.pyplot as plt
```

```python
def calculate_factorials(numbers):
    return {num: math.factorial(num) for num in numbers}
```

```python
numbers = list(map(int, input("Enter a list of numbers separated by spaces: ").split()))

factorials = calculate_factorials(numbers)

print("Factorials:")
for num, fact in factorials.items():
    print(f"{num}! = {fact}")
```

```python
plt.bar(factorials.keys(), factorials.values(), color='skyblue')
plt.yscale('log')  
plt.xlabel('Number')
plt.ylabel('Factorial (log scale)')
plt.title('Factorials of Numbers')
plt.show()
```
